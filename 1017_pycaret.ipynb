```python
# prompt: /content/NewsResult_20240717-20241017.xlsx 를 읽고 단어를 추출하여 contvector로 만들어줘

import pandas as pd
from konlpy.tag import Okt
from sklearn.feature_extraction.text import CountVectorizer

# 엑셀 파일 읽기
df = pd.read_excel('/content/NewsResult_20240717-20241017.xlsx')

# Okt 객체 생성
okt = Okt()

# 텍스트 데이터 전처리 및 명사 추출 함수 정의
def extract_nouns(text):
  if isinstance(text, str):  # 텍스트가 문자열인 경우에만 처리
    nouns = okt.nouns(text)
    return ' '.join(nouns)
  else:
    return ''

# '제목' 열의 텍스트 데이터에서 명사 추출
df['nouns'] = df['제목'].apply(extract_nouns)

# CountVectorizer를 사용하여 단어 벡터 생성
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['nouns'])
# 단어 사전 확인
print(vectorizer.vocabulary_)
```

```python
# prompt: X를 데이터프레임으로 컬럼명을 붙여서

X_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
X_df
```
