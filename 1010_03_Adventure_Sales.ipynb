## 1.  파일 불러오기

```python
# prompt: 노트북 셀내 여러 명령어의 출력을 다 출력해줘

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# prompt: https://github.com/microsoft/powerbi-desktop-samples/blob/main/AdventureWorks%20Sales%20Sample/AdventureWorks%20Sales.xlsx 파일은 여러 시트로 구성되어 있는데 모든 시트를 읽어줘

import pandas as pd

#!pip install pandas openpyxl


# Replace with the actual URL of your Excel file
url = "https://github.com/microsoft/powerbi-desktop-samples/raw/main/AdventureWorks%20Sales%20Sample/AdventureWorks%20Sales.xlsx"

try:
  xls = pd.ExcelFile(url)
  sheet_names = xls.sheet_names

  all_sheets_data = {}
  for sheet_name in sheet_names:
    df = pd.read_excel(url, sheet_name=sheet_name)
    df.head()
    all_sheets_data[sheet_name] = df
    print(f"Sheet '{sheet_name}' loaded successfully.")

except Exception as e:
  print(f"Error loading Excel file: {e}")
```

```python
path = '/content/drive/MyDrive/2024_2학기/빅데이터처리/data'
```

```python
# prompt: all_sheets_data안의 모든 데이터프레임을 path 에다 pickle 파일로 저장해

import pickle

for sheet_name, df in all_sheets_data.items():
  file_path = f"{path}/{sheet_name}.pkl"
  with open(file_path, 'wb') as f:
    pickle.dump(df, f)
  print(f"DataFrame '{sheet_name}' saved as pickle file at '{file_path}'.")
```

```python
def summary(df):
    tf = pd.DataFrame(
        {
        '데이터 개수' : df.shape[0],
        '데이터형태(dtypes)': df.dtypes,
        '고유값 수(nunique)': df.nunique(),
        '비 결측치 수(notnull)': df.notnull().sum(),
        '결측치 수(null)': df.isnull().sum(),
        '결측치 수(null)비율': df.isnull().sum() / df.shape[0] * 100,
        '최빈단어 5개': [df[col].value_counts().head(5) for col in df.columns]
    }
        )
    return tf

summary(Customer)
```

```python
# prompt: Customer, Sales를 CustomerKey를 공통키로 merge

merged_df = pd.merge(Customer, Sales, on='CustomerKey', how='inner')
merged_df
```

```python
# prompt: column_info_df 코드를 아주 간단하게 만들어줘

import pandas as pd

def column_info(df):
  return pd.DataFrame({
      col: {
          '고유값': df[col].nunique(),
          '결측치': df[col].isnull().sum(),
          '결측치비율': df[col].isnull().mean(),
          '최빈값(5개)': df[col].value_counts().head(5).to_dict()
      } for col in df.columns
  }).T

column_info(df)
```

```python
for sheet_name, df in all_sheets_data.items():
  print(f"Sheet: {sheet_name}")
  column_info(df)
  print('*'*150)
```
