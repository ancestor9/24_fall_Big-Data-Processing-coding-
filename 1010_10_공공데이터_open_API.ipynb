```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from tqdm import tqdm

df = pd.DataFrame()

for page_num in tqdm(range(10)):
    page = str(page_num)
    params ={'serviceKey' : service_key, 'pageNo' : '1', 'numOfRows' : page,  # 페이지 변동
             'LAWD_CD' : '11110', 'DEAL_YMD' : '201512' }
    print(f'다운로드하는 페이지는 {page}.')
    response = requests.get(url, params=params)
    data = xmltodict.parse(resp.content.decode('utf-8'))
    tf = pd.DataFrame(data['response']['body']['items']['item'])

    df = df.append(tf)
```
