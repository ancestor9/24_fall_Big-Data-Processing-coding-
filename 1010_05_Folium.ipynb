## **지도 시각화**

- 데이터 : https://www.data.go.kr/data/15094182/fileData.do#layer_data_infomation

```python
import pandas as pd
path = '/content/도로교통공단_자전거사고 다발지역 개별사고 정보_20201231.xlsx'
df = pd.read_excel(path)
df.head()
```

```python
df = pd.read_excel(path, usecols = [2, 3, 17, 18,])
df.head()
```

```python
df.info()
```

```python
df['지점명'].str.split(' ', expand=True)
```

```python
df['시도'] = df['지점명'].str.split(' ', n=2, expand=True)[0]
df['시군구'] = df['지점명'].str.split(' ', n=2, expand=True)[1]
df.head()
```

```python
# prompt: df['지점명']의  text에서 괄호안의 내용을 제거해다오

df['주소'] = df['지점명'].str.replace(r'\(.*\)', '', regex=True)
df.head()
```

```python
df['사상자수'] = df['사망자수'] + df['부상자수']
df.head()
```

### **행적구역별 위경도 정보 수집하기**

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
```

```python
geo_local.geocode('서울특별시 종로구 종로6가')
```

```python
geo_local.geocode('서울특별시 종로구 종로6가').latitude
```

```python
geo_local.geocode('서울특별시 종로구 종로6가').longitude
```

```python
lat = []
lon =[]
for col in df['주소'] :
    #print(col)
    lat.append(geo_local.geocode(col).latitude)
    lon.append(geo_local.geocode(col).longitude)
```

```python
# prompt: 주소를 기준으로 사상자수를 그룹핑해줘

grouped_df = df.groupby('주소')['사상자수'].sum().reset_index()
grouped_df
```

```python
import time

lat = []; lon =[]
for col in grouped_df['주소'] :
    print(col)
    lat.append(geo_local.geocode(col).latitude)
    lon.append(geo_local.geocode(col).longitude)
    time.sleep(1)
```

```python
grouped_df.isnull().sum()
```

```python
import time

lat = []; lon =[]
for col in grouped_df['주소'] :
    print(col)
    location = geo_local.geocode(col) # Get the location object
    if location: # Check if the location is found
        lat.append(location.latitude)
        lon.append(location.longitude)
    else:
        print(f"Could not find location for: {col}") # Print the address if not found
        lat.append(None)  # or some default value
        lon.append(None)  # or some default value
    time.sleep(1)
```

```python
import time

add = []; lat = []; lon =[]
for col in grouped_df['주소'].sample(30) :
    print(col)
    add.append(col)
    location = geo_local.geocode(col) # Get the location object
    if location: # Check if the location is found
        lat.append(location.latitude)
        lon.append(location.longitude)
    else:
        print(f"Could not find location for: {col}") # Print the address if not found
        lat.append(None)  # or some default value
        lon.append(None)  # or some default value
    time.sleep(1)
```

```python
# prompt: add, lat, lon을 모두 합쳐서 데이터프레임으로

import pandas as pd
new_df = pd.DataFrame({'주소': add, '위도': lat, '경도': lon})
print(new_df)
```

```python
# prompt: newdf 와 grouped_df를 merge

import pandas as pd
merged_df = pd.merge(grouped_df, new_df, on='주소', how='inner')
merged_df
```

```python
# prompt: merged_df의 행에 Nan이 하나라도 있으면 제거하하

merged_df_cleaned = merged_df.dropna()
print(merged_df_cleaned)
```

```python
import folium

m = folium.Map(location=[37.5758699,126.9811831], zoom_start=12)
folium.Marker(location=[37.5758699,126.9811831],
              popup='을지로',
              icon=folium.Icon(color='blue',icon='star')
              ).add_to(m)
m
```

```python
merged_df_cleaned
```

```python
# prompt: folium을 사용하여 주소에 해당되는 사상자수를 표시하는 코드

for index, row in merged_df_cleaned.iterrows():
  folium.CircleMarker(
      location=[row['위도'], row['경도']],
      radius=row['사상자수'] / 5,  # Adjust the radius scaling as needed
      popup=f"{row['주소']}: {row['사상자수']}명",
      color='red',
      fill=True,
      fill_color='red'
  ).add_to(m)

m
```
